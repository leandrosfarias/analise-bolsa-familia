## Contextualização

Cadastro único ou CadÚnico é um programa que visa a coleta de dados com o objetivo de identificar as famílias de baixa renda existentes no Brasil, com a finalidade de inclusão em  programas de assistência social e redistribuição de renda como o Programa Bolsa Família (PBF).

O Programa Bolsa Família (PBF) é um programa de transferência de renda direta que possui algumas condições para que as famílias cadastradas no CadÚnico possam ter acesso a recursos financeiros. Esse programa tem como objetivo melhorar a vida das famílias em situação de pobreza e extrema pobreza no Brasil. O PBF foi criado pelo Governo Federal, em 2003, por meio da Medida Provisória n° 132, que posteriormente foi convertida em Lei n° 10.836, de 9 de janeiro de 2004.

```python
def fun_grupos_tradicionais_especificos(value):
    if value == 000: return 'Nenhuma'
    elif value == 101: return 'Família Cigana'
    elif value == 201: return 'Família Extrativista'
    elif value == 202: return 'Família de Pescadores Artesanais'
    elif value == 203: return 'Família pertencente a Comunidade de Terreiro'
    elif value == 204: return 'Família Ribeirinha'
    elif value == 205: return 'Família de Agricultores Familiares'
    elif value == 301: return 'Família Assentada da Reforma Agrária'
    elif value == 302: return 'Família beneficiária do Programa Nacional de Crédito Fundiário'
    elif value == 303: return 'Família Acampada'
    elif value == 304: return 'Família Atingida por Empreendimentos de Infraestrutura'
    elif value == 305: return 'Família de Preso do Sistema Carcerário'
    elif value == 306: return 'Família de Catadores de Material Reciclável'
# Grupos_tradicionais_especificos
data_familia_ben_copy['Grupos_tradicionais_especificos'] = data_familia_ben_copy['ind_parc_mds_fam'].apply(fun_grupos_tradicionais_especificos)
# Familia_indigena
data_familia_ben_copy['cod_familia_indigena_fam'] = pd.cut(data_familia_ben['cod_familia_indigena_fam'],
                                                           bins=[1, 2, 3, float("inf")],
                                                           right=False,
                                                           labels=['Sim', 'Não'])
# Familia_quilombola
data_familia_ben_copy['ind_familia_quilombola_fam'] = pd.cut(data_familia_ben['ind_familia_quilombola_fam'],
                                                             bins=[1, 2, 3, float("inf")],
                                                             right=False,
                                                             labels=['Sim', 'Não'])
```
